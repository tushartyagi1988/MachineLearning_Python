## Creating Predictor and Traget Variable data frames

```python
df.columns
```

```python
#Creating a data frame of all predictor variables with excluding target variable and text columns
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
x.head()
```

```python
#Creating a data frame of target variable
y = df['Price']
```

```python
y.head()
```

## Splitting Training and Test data

```python
#Performing train_test_split
from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.40, random_state=101)
```

## Training Linear Regression Model

```python
#Training the model
from sklearn.linear_model import LinearRegression
```

```python
#Creating Linear Regression object
lm = LinearRegression()
```

```python
#Creating a fit or training the model
lm.fit(x_train,y_train)
```

## Interpreting Coefficients

```python
#Printing Intercept
print(lm.intercept_)
```

```python
#Checking coefficients
lm.coef_
```

```python
cdf = pd.DataFrame(lm.coef_,x.columns,columns=['Coeff'])
```

```python
cdf
```

This means that if we hold all other predictor variables constant, a 1 unit increase in Avg. Area Income is associated with $21.52 increase in price. Similar for other Predictor Variables as well

## Predictions

```python
predictions = lm.predict(x_test)
```

```python
#Predicted values of target variable(Price) based on x_test
predictions
```

```python
#Actual Target variable(Price)
y_test
```

## Visually Analyzing

```python
#Creating Scatterplot
plt.scatter(y_test,predictions)
```

```python
#Creating Histogram of distribution of residuals
sns.distplot((y_test-predictions))
```

If the residuals in the histogram look to be normally distributed, then we have done a good job and model is a correct choice for data

```python
from sklearn import metrics
```

```python
#Mean Absolute Error
metrics.mean_absolute_error(y_test,predictions)
```

```python
#Mean Squared Error
metrics.mean_squared_error(y_test,predictions)
```

```python
#Root Mean Squared Error
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```

```python
#Coefficient of Determination of R Square or accuracy of model
lm.score(x_test,y_test)
```

Explains how much of the variation in the dependent variable is explained by the model
